```python
# import

import yfinance as yf
import pandas as pd
```

```python
# fetch data

ticker = "^NSEI"

data = yf.download(ticker, start="2015-01-01", end="2024-12-31")

data = data[['Close']]
data.dropna(inplace=True)

data.head()
```

```python
## returns

data['returns'] = data['Close'].pct_change()
data.dropna(inplace=True)
```

```python
# save 

data.to_csv("option-greeks/data/nifty_underlying.csv")
```
