```python
# import from scripts

import sys
sys.path.append("..")

import numpy as np
from pricing.black_scholes import option_price, delta, gamma, vega, theta
  
```

```python
# option sample

S = 20000      # Current underlying price
K = 20000      # Strike price of the option
T = 30/365     # Time to expiry in years (30 days)
r = 0.06       # Risk-free interest rate (annual, continuous)
sigma = 0.20   # Implied volatility (annualized standard deviation)
```

```python
# call

print("CALL")
print("Price:", option_price(S,K,T,r,sigma,"call"))
print("Delta:", delta(S,K,T,r,sigma,"call"))
print("Gamma:", gamma(S,K,T,r,sigma))
print("Vega :", vega(S,K,T,r,sigma))
print("Theta:", theta(S,K,T,r,sigma,"call"))
```

```python
# put

print("PUT")
print("Price:", option_price(S,K,T,r,sigma,"put"))
print("Delta:", delta(S,K,T,r,sigma,"put"))
print("Gamma:", gamma(S,K,T,r,sigma))
print("Vega :", vega(S,K,T,r,sigma))
print("Theta:", theta(S,K,T,r,sigma,"put"))
```
